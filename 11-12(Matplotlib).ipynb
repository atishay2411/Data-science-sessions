```python
plt.figure()
x = np.arange(10)
y = x**2
z = x**3
plt.plot(y, 'o-', z, '-o')
plt.xlabel('Numbers')
plt.ylabel('Squares and Cubes')
plt.legend(['Squares', 'Cubes'])
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
plt.bar(x,y,width = 0.8)
plt.xticks(x)
plt.yticks(y)
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
z = x * 4
plt.bar(x,y, width = 0.4, color = 'green')
plt.bar(x,z, width = 0.4, bottom = x,  color = 'orange')
plt.xticks(x)
plt.yticks(range(0,50,5))
```

```python
plt.figure()
x = np.arange(10)
y = x*2
plt.bar(x,y,width = 0.4, color = 'yellow')
plt.bar(x+0.5, y*2, width = 0.4, color = 'green')
```
