```python
col1 = np.linspace(0,10,1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1" : col1, "C2" : col2})
df.head(10)
```

```python
plt.figure()
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = df.C1, y = df.C2)
```

```python
plt.figure()
plt.style.use('seaborn-darkgrid')
sns.lineplot(df.C1, df.C2)
```

```python
plt.figure(figsize = (12,8))
sns.set(rc = {'axes.facecolor' : 'blue', 'axes.grid' : True, 'xtick.labelsize' : 16, 'ytick.labelsize':16})
sns.lineplot(df.C1, df.C2, data = df, linewidth = 2.5)
sns.lineplot(df.C1, df.C2 + 1, data = df, linewidth = 2.5)
```

```python
plt.figure(figsize=(12,8))
sns.set(rc = {'axes.facecolor' : '#283747', 'axes.grid' : False, 'xtick.labelsize' : 16, 'ytick.labelsize':16})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 3.0, color = '#FF5722')
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 3.0, color = '#FFEB3B')
```

```python
plt.figure(figsize=(12,8))
sns.set(rc = {'axes.facecolor' : '#283747', 'axes.grid' : False, 'xtick.labelsize' : 16, 'ytick.labelsize':16})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 3.0, color = '#FF5722')
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 3.0, color = '#FFEB3B')
plt.title('Cool stuff with seaborn')
```

```python
plt.figure(figsize=(12,8))
sns.set(rc = {'axes.facecolor' : 'white', 'axes.grid' : False, 'xtick.labelsize' : 16, 'ytick.labelsize':16})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 3.0, color = '#FF5722', label = 'SinWave')
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 3.0, color = '#FFEB3B', label = 'SinWave + 1')
plt.title('Cool stuff with seaborn')
```

```python
auto = pd.read_csv('Automobile_data.csv')
```

```python
auto.head(10)
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x = 'wheel-base', y = 'engine-size', data = auto)
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x = 'wheel-base', y = 'engine-size', data = auto, linewidth = 5)
plt.title('Wheel-Base Vs Engine-Size')
```

```python
plt.figure(figsize = (20,6))
sns.lineplot(data = auto['wheel-base'], linewidth = 1.5, label = 'Wheel Base')
sns.lineplot(data = auto['engine-size'], linewidth = 1.5, label = 'Engine Size')
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = 'wheel-base', y = 'engine-size', hue = 'fuel-type',style = 'fuel-type', markers = True, data = auto, linewidth = 3)
```

```python
plt.figure(figsize=(14,7))
sns.set(rc={'xtick.labelsize':17,'ytick.labelsize':10,'axes.labelsize':15 , "axes.grid":False})
# Use "Pallete" to specify the colors to be used for different levels of the hue
sns.lineplot(x="wheel-base" , y="engine-size" , hue = "fuel-type" ,style="fuel-type" , dashes = False,palette = 'CMRmap',markers=["o", ">"] ,data=auto)
plt.show()
```

```python
plt.figure(figsize = (9,9))
sns.scatterplot(x = 'engine-size', y = 'wheel-base', data = auto, color = 'green')
```

```python
plt.figure(figsize = (9,9))
sns.scatterplot(x = 'engine-size', y = 'wheel-base', data = auto, color = 'green', marker = '>')
```

```python
plt.figure(figsize = (9,9))
sns.scatterplot(x = 'engine-size', y = 'wheel-base', data = auto, color = 'green', hue = 'fuel-type')
```

```python
plt.figure(figsize = (10,10))
plt.gcf().text(.5,.9,'Scatter Plot', fontsize = 40, color = 'Black', ha = 'center', va = 'center')
sns.scatterplot(x = 'engine-size', y = 'wheel-base', hue = 'fuel-type', style = 'num-of-cylinders', data = auto)
```

```python
plt.figure(figsize=(14,16))
sns.relplot(x="wheel-base" , y="engine-size" , hue="num-of-cylinders" ,kind='scatter',height = 9,aspect = 1.5, data = auto)
```

```python
plt.figure(figsize=(14,16))
sns.set(rc={'xtick.labelsize':12,'ytick.labelsize':12,'axes.labelsize':12})
sns.relplot(x="wheel-base" , y="engine-size" , hue="fuel-type" ,kind='scatter',palette=["#FF3333" ,"#00CC00"], height=8.5,aspect =1,data=auto)
plt.show()
```

```python
plt.figure(figsize=(14,20))
sns.set(rc={'xtick.labelsize':20,'ytick.labelsize':20,'axes.labelsize':20})
sns.relplot(x="wheel-base" , y="engine-size" , hue="fuel-type" ,kind='line',col = 'num-of-cylinders',col_wrap = 3, height=10,aspect =1,data=auto)
plt.show()
```

```python
sns.set(rc={'xtick.labelsize':12,'ytick.labelsize':12,'axes.labelsize':12})
sns.relplot(x="wheel-base", y="engine-size", hue="fuel-type",col="fuel-type", row="body-style", data=auto)
plt.show()
```
